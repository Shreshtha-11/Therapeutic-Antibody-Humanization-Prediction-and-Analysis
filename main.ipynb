```python
from collections import Counter

amino_acids = list("ACDEFGHIKLMNPQRSTVWY")

def aa_composition(seq):
    counts = Counter(seq)
    length = len(seq)

    return {
        aa: counts.get(aa, 0)/length
        for aa in amino_acids
    }
```

For a simple feature set:

20 amino acid frequencies + length

getting:

Accuracy = 99.1%

is unusually high.

This usually means one of three things:

Possibility 1 (Most Likely): Species Bias

Your non-human class is:

Mouse
Rat
Camel

Camel antibodies (VHH/nanobodies) have distinctive sequence characteristics compared to human antibodies.

The model may be learning:

"Is this camel?" rather than "Is this humanized?"

Possibility 2: Dataset Leakage

We need to check whether nearly identical sequences appear in both train and test sets.

Because we split after combining sequences, related clonotypes can end up in both sets.

Possibility 3: Real Biological Signal

Human and rodent germline repertoires are genuinely different.

But 99% with only amino-acid composition is still surprisingly high.

```python
from collections import Counter
import pandas as pd

amino_acids = list("ACDEFGHIKLMNPQRSTVWY")

def sequence_to_features(seq):

    seq = seq.upper().strip()

    counts = Counter(seq)
    length = len(seq)

    features = {
        aa: counts.get(aa, 0) / length
        for aa in amino_acids
    }

    return pd.DataFrame([features])
```

```python
def get_category(score):

    if score >= 90:
        return "Highly Human-Like"

    elif score >= 70:
        return "Moderately Human-Like"

    elif score >= 40:
        return "Partially Humanized"

    else:
        return "Non-Human"
```

```python
def get_confidence(score):

    distance = abs(score - 50)

    if distance >= 40:
        return "Very High"

    elif distance >= 25:
        return "High"

    elif distance >= 10:
        return "Moderate"

    return "Low"
```

```python
def analyze_antibody(sequence):

    feat = sequence_to_features(sequence)

    prob = model2.predict_proba(feat)[0][1]

    score = round(prob * 100, 2)

    return {
        "Human-Likeness Score": score,
        "Category": get_category(score),
        "Confidence": get_confidence(score)
    }
```

Upgrade 3: Validation Layer

```python
VALID_AA = set(amino_acids)

def validate_sequence(seq):

    seq = seq.upper().strip()

    if len(seq) < 80:
        return False, "Sequence too short"

    if not set(seq).issubset(
        VALID_AA
    ):
        return False, "Invalid amino acids"

    return True, "Valid"
```

Upgrade 4 Mutation Simulation Engine

```python
def mutate_sequence(seq, position, new_aa):

    seq = list(seq)

    seq[position] = new_aa

    return "".join(seq)
```

```python
mutate_sequence(
    "ACDEFG",
    2,
    "W"
)
```

```python
def get_score(sequence):

    feat = sequence_to_features(sequence)

    prob = model2.predict_proba(feat)[0][1]

    return prob * 100
```

```python
AMINO_ACIDS = list(
    "ACDEFGHIKLMNPQRSTVWY"
)
```

```python
def best_single_mutations(
    sequence,
    top_n=10
):

    original_score = get_score(sequence)

    candidates = []

    for pos in range(len(sequence)):

        original_aa = sequence[pos]

        for aa in AMINO_ACIDS:

            if aa == original_aa:
                continue

            mutated = mutate_sequence(
                sequence,
                pos,
                aa
            )

            score = get_score(mutated)

            improvement = (
                score -
                original_score
            )

            if improvement > 0:

                candidates.append({

                    "position":
                        pos + 1,

                    "mutation":
                        f"{original_aa}{pos+1}{aa}",

                    "new_score":
                        round(score,2),

                    "gain":
                        round(improvement,2)
                })

    candidates = sorted(
        candidates,
        key=lambda x: x["gain"],
        reverse=True
    )

    return candidates[:top_n]
```

```python
mouse_seq = combined_df[
    combined_df["species"]=="mouse"].iloc[0]["sequence"]
```

```python
results = best_single_mutations(
    mouse_seq,
    top_n=10
)

results[:10]
```
